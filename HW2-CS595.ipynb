## Data set
Statlog (German Credit Data) Data Set. <https://archive.ics.uci.edu/ml/datasets/statlog+(german+credit+data)>

This dataset classifies people as with good or bad credit based on a set of attributes. The features include Status of existing checking account, Duration in month, Credit history , Purpose, Credit amount, Personal status and sex, Property, Age, Housing, Job, etc. The categorical attributes are numerized.

```python
#load data (pre-processed by myself by removing extra whitespaces and converting to csv)
import csv
import numpy as np
reader=csv.reader(open("GermanCreditData.csv","rt"),delimiter=',')
data=list(reader)
Xy=np.array(data).astype('int')#load feature data
X=Xy[:,0:24]
y=Xy[:,[24]]
print('Data shape:\n%d features, %d instances' % (len(X[0]),len(X)))
print('\nHead 5 lines of X:')
print(X[:5])
print('\nHead 5 lines of y (1 = Good credit, 2 = Bad credit):')
print(y[:5])
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.333)
```

## Decision tree based classification
### max_depth=1

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
clf = DecisionTreeClassifier(max_depth=1)
clf.fit(X_train,y_train)
train_pred=clf.predict(X_train)
test_pred=clf.predict(X_test)
print("Train accuracy = %f" % accuracy_score(y_train,train_pred))
print("Test accuracy = %f" % accuracy_score(y_test,test_pred))
```

```python
import graphviz
dot_data=export_graphviz(clf,out_file=None)
graph = graphviz.Source(dot_data) 
graph
```

## Discussion

```python
%matplotlib inline
#plot the change of accuracy wrt max_depth
import matplotlib.pyplot as plt
maxdepth=range(1,11)
train_acc=[0.693910,0.714744,0.754808,0.772436,0.802885,0.852564,0.895833,0.923077,0.945513,0.971154]
test_acc=[0.711538,0.714744,0.727564,0.740385,0.692308,0.701923,0.701923,0.721154,0.708333,0.714744]
plt.plot(maxdepth,train_acc)
plt.plot(maxdepth,test_acc)
plt.xlabel("max_depth of decision tree classifier")
plt.ylabel("Accuracy")
plt.legend(["Train acc","Test acc"],loc='upper left')
plt.show()
```

**Observation**: According to the figure above, we find that the training accuracy keeps increasing when we increase max_depth, i.e. allowing taller/deeper trees. However, the test accuracy is not monetone increasing. It increases until max_depth=4 and reaches the maximum there. Decision trees with depth>4 does not work well on the test set because their structures are very complex and they tend to overfit the training set.

**Conclusion**: It's important to keep the model simple. Simple chasing for perfect accuracy on the training set usually cause overfitting.
